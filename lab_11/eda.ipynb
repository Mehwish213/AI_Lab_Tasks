```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('cleaned_data.csv')
print(df)
```

<div class="alert alert-block alert-info">
    5. <b>Train - Test Splitting</b>
</div>

```python
df.columns
```

```python
X = df[['open','high','low','volume','year','month','day','day_of_week']]
y = df['close']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape, train_y.shape)
print(test_X.shape, test_y.shape)
```

<div class="alert alert-block alert-info">
    6. <b>Applying Classifier / Model Training</b>
</div>

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_X, train_y)

print(model)
```

```python
model_pred = model.predict(test_X)
print(model_pred)
```

```python
from sklearn.metrics import r2_score

model_pred = model.predict(test_X)
model_acc = r2_score(test_y, model_pred)
print(round(model_acc, 3) * 100, "%")
```

```python
import pickle
pickle.dump(model, open('model.pkl', 'wb'))
```
