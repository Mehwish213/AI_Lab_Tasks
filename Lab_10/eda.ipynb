```python
df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]
print("\n Cleaned Column Names:")
print(df.columns)
```

```python
print("\n Checking for missing values before handling:")
print(df.isnull().sum())
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ['float64', 'int64']:
            df[col].fillna(df[col].mean(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)
print("\n Missing values handled successfully!")
print(df.isnull().sum())
```

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date']) 
print("\n 'date' converted to datetime format successfully.")
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek

print("\n Extracted Year, Month, Day, and Day of Week columns.")
```

```python
numeric_cols = ['open', 'high', 'low', 'close', 'adj_close', 'volume']
for col in numeric_cols:
    df[col] = df[col].astype(float).astype(int)

print("\n Numeric data converted to integers where appropriate!")
```

```python
duplicates = df.duplicated().sum()
if duplicates > 0:
    df = df.drop_duplicates()
print(f"\n Removed {duplicates} duplicate rows.")
```

```python
print("\n Summary Statistics:")
print(df.describe())
```

```python
print("\n Final Preprocessed Dataset:")
print(df.head())
```

```python
df.to_csv("cleaned_data.csv", index=False)
print("\n Cleaned dataset saved as 'cleaned_data.csv' successfully!")
```
